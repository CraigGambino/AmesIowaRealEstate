Linear Regression with Sklearn

```python
ss = StandardScaler()
lasso = Lasso()
pipe = Pipeline([
    ('ss', ss),
    ('lasso', lasso)
])

parameters = {
    'lasso__alpha': np.arange(0,100,5)
    
}

scoring = ['explained_variance',
    'neg_mean_absolute_error',
    'r2']
```

```python
for score in scoring:
    gs = GridSearchCV(pipe, param_grid=parameters, n_jobs=6, scoring=score)
    gs.fit(X_train, y_train)

    print(gs.best_score_)
    print(gs.best_params_)
```

```python
llr = Lasso(alpha=gs.best_params_['lasso__alpha'])
llr.fit(X_train, y_train)
feat_coef = dict(zip(features, llr.coef_.tolist()))
```
